```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams.update({'font.size': 18})
plt.rcParams['figure.figsize'] = [6, 6]
%matplotlib inline

# arrays
x = np.linspace(-np.pi, np.pi, num=100)
y = np.linspace(-np.pi, np.pi, num=100)
X = np.tensordot(x, np.ones_like(y), axes=0)
Y = np.tensordot(np.ones_like(x), y, axes=0)

def vorticity(X,Y,n,A,phi):
    return n * A * (np.cos(n*X+phi) + np.sin(n*Y+phi))

num = 5
A = [1/(n**2.0) for n in range(1,num+1)]
phi = np.pi * np.random.randn(num)

F = sum([vorticity(X,Y,idx,A[idx],phi[idx]) for idx in range(num)])

cb = np.linspace(np.amin(F), np.amax(F), num=100)

# Figure
plt.figure()
plt.contour(X, Y, F, 0) # cb)
plt.show()
```
